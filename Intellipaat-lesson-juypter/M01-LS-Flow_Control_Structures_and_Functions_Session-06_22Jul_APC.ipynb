## Why use `for` loop?

Let’s see the use **`for`** loop in Python.

* **Definite Iteration:** When we know how many times we wanted to run a loop, then we use count-controlled loops such as **`for`** loops. It is also known as definite iteration. For example, Calculate the percentage of 50 students. here we know we need to iterate a loop 50 times (1 iteration for each student).
* **Reduces the code’s complexity:** Loop repeats a specific block of code a fixed number of times. It reduces the repetition of lines of code, thus reducing the complexity of the code. Using **`for`** loops and while loops we can automate and repeat tasks in an efficient manner.
* **Loop through sequences:** used for iterating over lists, strings, tuples, dictionaries, etc., and perform various operations on it, based on the conditions specified by the user.

## Break, Continue and Pass

### Break

- it will stop execution of the `for` loop when executed (breaks & come out of the loop)
- break terminates the current loop given the condition satisfies if used with conditional statements

```python
nums = [10,20,30,40,60,150,45,65,33,105,90,99]
```

#### Print all the numbers in the above list below 100

```python
nums = [10,20,30,40,60,150,45,65,33,105,90,99]

for i in nums:
    if i>=100:
        break
    print('Current number is:', i)
    
    
print('Out of the for loop')
```

**Similar to `do-while` in C**

### Continue

- it is used to stop/skip the block of the code in the `for` loop for the current iteration only and continue with the next iteration

```python
nums = [10,20,30,40,60,150,45,65,33,105,90,99]

for i in nums:
    if i>=100:
        continue
    print('Current number is:', i)
    
    
print('Out of the for loop')
```

### Pass

- is a null statment, i.e., nothing happens when the statement is executed

```python
nums = [10,20,30,40,60,150,45,65,33,105,90,99]

for i in nums:
    if i>=100:
        pass
    print('Current number is:', i)
    
    
print('Out of the for loop')
```

## Function

### What is a function in Python?

In Python, a **function is a block of organized, reusable (DRY- Don’t Repeat Yourself) code with a name** that is used to perform a single, specific task. It can take arguments and returns the value.

Functions help break our program into smaller and modular chunks. As our program grows larger and larger, functions make it more organized and manageable.

Furthermore, it improves efficiency and reduces errors because of the reusability of a code.

### Types of Functions

Python support two types of functions
1. Built-in function 
2. User-defined function

#### Built-in function

- Functions which come along with Python itself are called a built-in function or predefined function
- Some of them are `range()`, `print()`, `input()`, `len()`

#### User-defined function

- Functions which are created by the programmer (you guys) according to the specific requirements

#### Syntax

`def`: keyword used to declare a funciton

`function_name`: any name given to the function

`arguments/parameters`: any value passed to the function

`return`: return value from a function (optional)

### Q. Write a function which swaps the value of two variables
`Given value is: (6,9), the swapped value is (9,6)`

`Expected Output`

    Swapped value of x is:9
    Swapped value of y is:6

```python
def swap_values(x,y):
    temp = x #storing the value of x in 'temp' variable
    x=y #assigning y to x
    y=temp #lastly assign temp to y which had original value of x
    
    print('Swapped value of x is:', x)
    print('Swapped value of y is:', y)
```

### Q. Write a function to do certain operations on the two numbers

- Addition: add
- Subtraction: sub
- Multiplication: mul
- Division: div
- Floor Division: fdiv
- Modulus: mod
- Exponentiation/Power : exp

Select the operation by passing a parameter .Also make sure that the parameter is optional in nature with the **default value 'add'**

`Hint: For optional parameter:` **operation**, default operation is `add` 

```python
def math_operate(num1, num2, operation='add'):
    if operation == 'add':
        result = num1 + num2
        
    elif operation == 'sub':
        result = num1 - num2
    
    elif operation == 'mul':
        result = num1* num2
        
    elif operation == 'div':
        result = num1/num2
    
    elif operation == 'fdiv':
        result = num1//num2
        
    elif operation == 'mod':
        result = num1%num2
        
    elif operation == 'exp':
        result = num1**num2
    else:
        return "Invalid Operation"
        
    return result
   
```

```python
math_operate(0,0, 'div')
```
